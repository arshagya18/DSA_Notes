### Understanding Python Dictionaries (Hashmaps)

Let's clarify how Python dictionaries handle duplicate keys, duplicate values, and various access methods.

#### 1. Can Python Dictionaries have Duplicate Keys?

No, Python dictionaries **cannot** have duplicate keys. Keys in a dictionary must be unique. If you try to assign a value to an existing key, the old value associated with that key will be overwritten (updated) with the new value.

**Example:**
```python
my_dict = {'a': 1, 'b': 2, 'c': 3}
print(f"Initial dictionary: {my_dict}")

# Trying to add a duplicate key 'a'
my_dict['a'] = 10 # This updates the value for key 'a', it does not create a new entry
print(f"Dictionary after adding a duplicate key 'a': {my_dict}")

my_dict['b'] = 20 # Updating another key
print(f"Dictionary after updating 'b': {my_dict}")
```

#### 2. Can Python Dictionaries have Duplicate Values?

Yes, Python dictionaries **can** have duplicate values. There's no restriction on values being unique; only keys must be unique.

**Example:**
```python
my_dict = {'apple': 10, 'banana': 20, 'cherry': 10, 'date': 20}
print(f"Dictionary with duplicate values: {my_dict}")
# Here, 'apple' and 'cherry' have the same value (10).
# Also, 'banana' and 'date' have the same value (20).
```

#### 3. How to Access a Value Using its Key?

You can access a value using its key in two primary ways:

*   **Using square brackets `[]`**: This is the most common method. If the key does not exist, it will raise a `KeyError`.
*   **Using the `get()` method**: This method allows you to specify a default value to return if the key is not found, preventing a `KeyError`.

**Example:**
```python
my_dict = {'name': 'Alice', 'age': 30, 'city': 'New York'}

# Accessing using square brackets
print(f"Name: {my_dict['name']}")

# Accessing using .get() method
print(f"Age: {my_dict.get('age')}")

# Accessing a non-existent key with square brackets (will raise KeyError if uncommented)
# print(my_dict['country'])

# Accessing a non-existent key with .get() (returns None by default)
print(f"Country (using get without default): {my_dict.get('country')}")

# Accessing a non-existent key with .get() (with a specified default value)
print(f"Country (using get with default): {my_dict.get('country', 'Unknown')}")
```

#### 4. How to Access a Key Using its Value?

Unlike accessing values by keys, there is no direct or built-in method to efficiently find a key by its value in Python dictionaries. This is because values are not required to be unique, and searching by value can be an `O(n)` operation (linear time), as the dictionary would need to iterate through all items. If you have duplicate values, there might be multiple keys associated with that value.

To find a key by its value, you typically need to iterate through the dictionary's items.

**Example:**
```python
my_dict = {'item1': 'apple', 'item2': 'banana', 'item3': 'apple', 'item4': 'cherry'}
search_value = 'apple'

found_keys = []
for key, value in my_dict.items():
  if value == search_value:
    found_keys.append(key)

print(f"Dictionary: {my_dict}")
print(f"Keys for value '{search_value}': {found_keys}")

search_value_unique = 'banana'
found_key_unique = None
for key, value in my_dict.items():
  if value == search_value_unique:
    found_key_unique = key
    break # Assuming we only need the first one
print(f"Key for value '{search_value_unique}': {found_key_unique}")

# If you frequently need to look up keys by values, you might consider creating a 'reverse' dictionary
# (mapping values to lists of keys), but this needs to be maintained if the original dictionary changes.
```

````python
from collections import defaultdict

# Using the example dictionary from the previous explanation
my_dict = {'item1': 'apple', 'item2': 'banana', 'item3': 'apple', 'item4': 'cherry'}

# Create a defaultdict to store the reverse mapping
reverse_dict = defaultdict(list)

# Iterate through the original dictionary to populate the reverse dictionary
for key, value in my_dict.items():
    reverse_dict[value].append(key)

print(f"Original dictionary: {my_dict}")
print(f"Reverse dictionary (value to keys): {dict(reverse_dict)}")

# Now you can easily look up keys by value:
search_value_1 = 'apple'
print(f"Keys for '{search_value_1}': {reverse_dict[search_value_1]}")

search_value_2 = 'banana'
print(f"Keys for '{search_value_2}': {reverse_dict[search_value_2]}")

search_value_3 = 'grape'
print(f"Keys for non-existent value '{search_value_3}': {reverse_dict[search_value_3]}") # Returns an empty list if not found
````

### Explanation of the Reverse Dictionary Code:

1.  **`from collections import defaultdict`**: We import `defaultdict` from the `collections` module. This special dictionary subclass allows you to specify a default value factory (e.g., `list`) for keys that haven't been seen yet. When you try to access a key that doesn't exist, it automatically creates a new list for it, preventing `KeyError`.

2.  **`reverse_dict = defaultdict(list)`**: This initializes our reverse dictionary. When we try to access `reverse_dict[some_value]` for the first time, if `some_value` isn't already a key, `defaultdict` will automatically create an empty list (`[]`) as its value.

3.  **`for key, value in my_dict.items():`**: We iterate through each key-value pair in the original `my_dict`.

4.  **`reverse_dict[value].append(key)`**: For each pair, we use the `value` from the original dictionary as the key in our `reverse_dict`. We then `append` the `key` from the original dictionary to the list associated with that `value` in `reverse_dict`.

    *   If `value` is seen for the first time, an empty list is automatically created for it, and the `key` is appended.
    *   If `value` has been seen before, the `key` is simply appended to the existing list.

5.  **`print(f"Reverse dictionary (value to keys): {dict(reverse_dict)}")`**: We convert the `defaultdict` back to a regular `dict` for printing to show its final structure clearly.

This approach is efficient for creating the reverse mapping and allows you to easily retrieve all keys associated with a specific value, even if there are multiple.

### Creating a Hash Map (Dictionary) from a Given Array

There are several scenarios for creating a dictionary from an 'array' (which usually means a list in Python). Let's look at the most common ones:

#### Scenario 1: Array of Key-Value Pairs (Tuples or Lists)

If your array is a list of tuples or lists, where each inner element represents a `(key, value)` pair, you can directly pass it to the `dict()` constructor. This is generally the most straightforward and Pythonic way.


````python
# Using a list of tuples
list_of_tuples = [('name', 'Alice'), ('age', 30), ('city', 'New York')]
my_dict_from_tuples = dict(list_of_tuples)
print(f"Dictionary from list of tuples: {my_dict_from_tuples}")

# Using a list of lists
list_of_lists = [['country', 'USA'], ['language', 'English']]
my_dict_from_lists = dict(list_of_lists)
print(f"Dictionary from list of lists: {my_dict_from_lists}")
````

#### Scenario 2: Two Separate Arrays (One for Keys, One for Values)

If you have one list for keys and another list for values, you can use the `zip()` function to combine them into an iterator of key-value pairs, and then pass that to the `dict()` constructor.

**Important:** For this to work correctly, both lists should have the same length.

````python
keys = ['product_id', 'product_name', 'price']
values = [101, 'Laptop', 1200.00]

my_dict_from_two_lists = dict(zip(keys, values))
print(f"Dictionary from two separate lists: {my_dict_from_two_lists}")

# What happens if lengths are different?
keys_short = ['a', 'b']
values_long = [1, 2, 3]
my_dict_short_keys = dict(zip(keys_short, values_long))
print(f"Dictionary with shorter keys list: {my_dict_short_keys}") # 'c' and 3 are ignored

keys_long = ['x', 'y', 'z']
values_short = [10, 20]
my_dict_short_values = dict(zip(keys_long, values_short))
print(f"Dictionary with shorter values list: {my_dict_short_values}") # 'z' is ignored
````

#### Scenario 3: Creating a Dictionary with Default Values

If you have a list of keys and want to assign a default value to each, or if you want to use a function to generate values, you can use `dict.fromkeys()` or a dictionary comprehension.


````python
# Using dict.fromkeys() with a default value
keys_list = ['item_a', 'item_b', 'item_c']
default_value = 0
my_dict_default = dict.fromkeys(keys_list, default_value)
print(f"Dictionary with default value: {my_dict_default}")

# Using a dictionary comprehension to generate values (e.g., squares of numbers)
numbers = [1, 2, 3, 4]
my_dict_comprehension = {num: num**2 for num in numbers}
print(f"Dictionary using comprehension (squares): {my_dict_comprehension}")

# Another comprehension example: mapping strings to their lengths
words = ['apple', 'banana', 'cat']
my_dict_lengths = {word: len(word) for word in words}
print(f"Dictionary using comprehension (lengths): {my_dict_lengths}")
````

### Creating a Frequency Map (Counter) from an Array

A frequency map (or counter) is a dictionary-like object where keys are the items from your array, and values are the counts of how many times each item appears. Here are the best ways to create one in Python:

#### 1. Using `collections.Counter` (Most Pythonic and Recommended)

The `collections` module provides a `Counter` class that is a subclass of `dict` specifically for counting hashable objects. It's the most efficient and readable way to create a frequency map.


````python
from collections import Counter

my_array = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 'apple', 'banana', 'apple', 'orange']

# Create a frequency map using Counter
frequency_map_counter = Counter(my_array)

print(f"Original array: {my_array}")
print(f"Frequency map (using Counter): {frequency_map_counter}")

# Accessing counts:
print(f"Count of 'apple': {frequency_map_counter['apple']}")
print(f"Count of 3: {frequency_map_counter[3]}")
print(f"Count of 'grape' (non-existent): {frequency_map_counter['grape']}") # Returns 0 for non-existent items

# Common operations:
print(f"Most common 2 items: {frequency_map_counter.most_common(2)}")
````

#### 2. Using `collections.defaultdict`

The `defaultdict` from the `collections` module is another elegant way. You initialize it with `int` as the default factory, so when a new key is accessed, it automatically gets a default value of `0`.


````python
from collections import defaultdict

my_array = ['red', 'blue', 'red', 'green', 'blue', 'red']

# Create a frequency map using defaultdict
frequency_map_defaultdict = defaultdict(int)

for item in my_array:
    frequency_map_defaultdict[item] += 1

print(f"Original array: {my_array}")
print(f"Frequency map (using defaultdict): {dict(frequency_map_defaultdict)}")

# Accessing counts:
print(f"Count of 'red': {frequency_map_defaultdict['red']}")
print(f"Count of 'yellow' (non-existent): {frequency_map_defaultdict['yellow']}") # Returns 0
````

#### 3. Using a Simple Loop with a Regular Dictionary

You can also build a frequency map using a standard Python dictionary and a `for` loop. This method requires checking if a key already exists before incrementing its count or setting it to 1.


````python
my_array = ['cat', 'dog', 'cat', 'bird', 'dog', 'cat']

# Create a frequency map using a simple loop
frequency_map_loop = {}

for item in my_array:
    if item in frequency_map_loop:
        frequency_map_loop[item] += 1
    else:
        frequency_map_loop[item] = 1

print(f"Original array: {my_array}")
print(f"Frequency map (using simple loop): {frequency_map_loop}")

# Alternatively, using dict.get() for a slightly cleaner loop:
frequency_map_get = {}
for item in my_array:
    frequency_map_get[item] = frequency_map_get.get(item, 0) + 1

print(f"Frequency map (using dict.get()): {frequency_map_get}")
````

### Understanding `enumerate()`

The `enumerate()` function is commonly used when you need to iterate through a sequence (like a list) and also keep track of the index of the current item.

It returns pairs of `(index, value)` for each item in the iterable.

````python
nums = [19, 4, 6, 7, 8, 8, 87, 5, 4, 3]

print(f"Original list: {nums}")

# When you simply print enumerate(nums), you get the enumerate object itself.
# To see its contents, you usually convert it to a list or iterate over it.
print(f"enumerate(nums) object: {enumerate(nums)}")
print(f"Contents of enumerate(nums) as a list: {list(enumerate(nums))}")

print("\nIterating through nums with enumerate():")
for index, value in enumerate(nums):
    print(f"Index: {index}, Value: {value}")

# You can also start the count from a different number
print("\nIterating through nums with enumerate() starting count from 1:")
for index, value in enumerate(nums, start=1):
    print(f"Index: {index}, Value: {value}")
````

## Notes: 'Contains Duplicate' Problem - Optimized Solution

**Problem Statement:** Given an integer array `nums`, return `true` if any value appears at least twice in the array, and return `false` if every element is distinct.

**Original Solution (using Hash Map / Frequency Counter):**
```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        hash_map = {}

        for num in nums:
            hash_map[num] = hash_map.get(num, 0) + 1
            if hash_map[num] > 1:
                return True
        return False
```

**Optimized Solution (using Set):**

The most Pythonic and often fastest way to solve this is by comparing the length of the original list to the length of a set created from the list. Sets inherently store only unique elements, so if the lengths differ, duplicates must have been present.

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        return len(nums) != len(set(nums))
```

**Explanation of the Optimized Solution:**

1.  **`set(nums)`**: This creates a new set from the elements of the `nums` list. When a list is converted to a set, any duplicate elements are automatically removed, leaving only unique values.
2.  **`len(set(nums))`**: This gives you the count of unique elements in the original `nums` list.
3.  **`len(nums)`**: This gives you the total count of elements (including duplicates) in the original `nums` list.
4.  **`len(nums) != len(set(nums))`**: If the number of unique elements is *not equal* to the total number of elements, it means that some elements were removed during the set conversion, which implies duplicates were present. In this case, the expression evaluates to `True`.

**Why this is 'better':**

*   **Conciseness and Readability:** It's a single, very clear line of code that expresses the intent directly.
*   **Efficiency:** Both the set creation (`set(nums)`) and length comparison are highly optimized in Python's C implementation. On average, creating a set from a list takes O(n) time complexity, and checking the length is O(1). The space complexity is also O(n) in the worst case (all unique elements), similar to the hash map approach. For this specific problem, it's often the most performant and idiomatic Python solution.

## Notes: 'Valid Anagram' Problem - Optimized Solution

**Problem Statement:** Given two strings `s` and `t`, return `true` if `t` is an anagram of `s`, and `false` otherwise.

An **anagram** is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

**Your Current Solution (using Two Hash Maps):**
This approach correctly counts the frequency of each character in both strings and then compares the resulting frequency maps. If the character counts are identical for both strings, they are anagrams.

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        # Initial check: Anagrams must have the same length
        if len(s) != len(t):
            return False

        hashmap1 = {}
        hashmap2 = {}

        for char_s in s:
            hashmap1[char_s] = hashmap1.get(char_s, 0) + 1
        for char_t in t:
            hashmap2[char_t] = hashmap2.get(char_t, 0) + 1

        return hashmap1 == hashmap2
```

**Optimized / Pythonic Solution (using `collections.Counter`):**

The `collections.Counter` class is specifically designed for counting hashable objects. It's a subclass of `dict` that makes frequency counting very efficient and readable. Since a `Counter` object is essentially a dictionary, two `Counter` objects can be directly compared using `==` to check if they have the same counts for the same elements.

```python
from collections import Counter

class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        # Initial check: Anagrams must have the same length
        if len(s) != len(t):
            return False
            
        return Counter(s) == Counter(t)
```

**Explanation of the Optimized Solution:**

1.  **`from collections import Counter`**: Imports the `Counter` class.
2.  **`if len(s) != len(t): return False`**: This is a crucial early exit condition. If two strings are to be anagrams, they *must* have the same length. This check can save computation for many non-anagram pairs.
3.  **`Counter(s)`**: Creates a `Counter` object for string `s`. This object will store character counts (e.g., `Counter({'a': 3, 'n': 1, 'g': 1, 'r': 1, 'm': 1})` for "anagram").
4.  **`Counter(t)`**: Creates a `Counter` object for string `t` in the same way.
5.  **`Counter(s) == Counter(t)`**: Directly compares the two `Counter` objects. This comparison returns `True` if both `Counter` objects have the exact same elements with the exact same frequencies, and `False` otherwise.

**Why this is 'better':**

*   **Conciseness and Readability:** It's a single, highly readable line of code that clearly conveys the intent of comparing character frequencies.
*   **Efficiency:** `collections.Counter` is implemented in C for performance, making it very efficient for this task. The time complexity is O(N) (where N is the length of the strings) because it iterates through each string once to build the counts. The space complexity is O(alphabet_size) as it stores counts for unique characters.

### Follow-up: What if the inputs contain Unicode characters?

Both the hashmap (dictionary) approach and the `collections.Counter` approach in Python handle Unicode characters correctly without any modifications. Python strings are inherently Unicode, and characters like 'é', '你好', or '😊' are treated as individual characters.

*   When iterating through a Python string with a `for` loop (e.g., `for char in s:`), `char` will correctly be each Unicode character.
*   Dictionaries and `Counter` objects will store these Unicode characters as keys and correctly count their frequencies.

Therefore, your solutions are already robust for Unicode inputs!

## Notes: 'First Unique Character in a String' Problem

**Problem Statement:** Given a string `s`, find the first non-repeating character in it and return its index. If it does not exist, return -1.

**Your Current Solution (using a Dictionary/Hash Map):**
This approach correctly counts the frequency of each character in the string using a dictionary. Then, it iterates through the string again to find the first character whose count in the dictionary is 1.

```python
class Solution:
    def firstUniqChar(self, s: str) -> int:
        char_counts = {}
        # First pass: Populate character counts
        for char in s:
            char_counts[char] = char_counts.get(char, 0) + 1

        # Second pass: Find the first character with a count of 1
        for i, char in enumerate(s):
            if char_counts[char] == 1:
                return i
        return -1 # No unique character found
```

**Optimized / Pythonic Solution (using `collections.Counter`):**

The `collections.Counter` class is ideal for frequency counting and can make the code more concise while maintaining efficiency.

```python
from collections import Counter

class Solution:
    def firstUniqChar(self, s: str) -> int:
        # First pass: Populate character counts using Counter
        char_counts = Counter(s)

        # Second pass: Find the first character with a count of 1
        for i, char in enumerate(s):
            if char_counts[char] == 1:
                return i
        return -1 # No unique character found
```

**Explanation of Both Solutions:**

1.  **First Pass (Counting Frequencies):**
    *   Both solutions first iterate through the input string `s` to build a frequency map of all characters. This takes O(N) time, where N is the length of the string.
    *   Your solution uses a standard Python dictionary and the `dict.get(key, default_value)` method, which is efficient.
    *   The `collections.Counter` solution achieves the same result in a single, more concise line: `char_counts = Counter(s)`. `Counter` is optimized in C and is generally the preferred Pythonic way for frequency counting.

2.  **Second Pass (Finding the First Unique):**
    *   Both solutions then iterate through the string `s` *again*, this time using `enumerate` to get both the index (`i`) and the character (`char`).
    *   For each character, they check its count in the `char_counts` map. If the count is 1, it means this is the first non-repeating character encountered, and its index `i` is returned.
    *   If the loop finishes without finding any character with a count of 1, it means all characters repeat, so -1 is returned.

**Time and Space Complexity:**

*   **Time Complexity:** O(N), where N is the length of the string `s`. This is because we iterate through the string twice: once to build the frequency map and once to find the first unique character.
*   **Space Complexity:** O(alphabet_size). In the worst case, if all characters are unique, the dictionary (or `Counter`) will store an entry for each unique character. For lowercase English letters, this is a constant 26 characters. **Problem Statement:** Given two strings `ransomNote` and `magazine`, return `true` if `ransomNote` can be constructed by using the letters from `magazine` and `false` otherwise.

This solution correctly counts the frequency of each character in both `ransomNote` and `magazine`. It then iterates through the `ransomNote` characters (by effectively checking the counts in `ran`) and verifies if `magazine` (`mag`) contains enough of each required character.

```python
class Solution:
```
```python
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
```

```python
        # p2: pointer for the end of nums2

        nums1[:p2 + 1] = nums2[:p2 + 1]
```

## Notes: 'Majority Element' Problem

**Problem Statement:** Given an array `nums` of size `n`, return the majority element. The majority element is the element that appears more than `⌊n / 2⌋` times. You may assume that the majority element always exists in the array.

**Your Solution (using `collections.Counter`):**
Your solution correctly identifies the most frequent element by iterating through the array and comparing counts from a frequency map.

**Ideal Solution (Boyer-Moore Voting Algorithm):**

```python
class Solution:

        candidate = None

                candidate = num
            count += (1 if num == candidate else -1)

        return candidate
```

## Notes: 'Missing Number' Problem

**Problem Statement:** Given an array `nums` containing `n` distinct numbers in the range `[0, n]`, return the only number in the range that is missing from the array.

**Your Solution (using `set`):**
Your approach converts the list to a set for O(1) lookups and then checks every number in the range `[0, n]` until the missing one is found.

```python

    def missingNumber(self, nums: List[int]) -> int:
        nums_set = set(nums)
        for i in range(len(nums)+1):
            if i not in nums_set:
                return i
```

**Ideal Solution (Mathematical / Gauss' Formula):**
The sum of the first `n` natural numbers is given by the formula: `n * (n + 1) // 2`. If we calculate the expected sum and subtract the actual sum of the elements in the array, the difference is the missing number.

```python
class Solution:
    def missingNumber(self, nums: List[int]) -> int:
        n = len(nums)
        expected_sum = n * (n + 1) // 2
        actual_sum = sum(nums)
        return expected_sum - actual_sum
```

**Alternative Solution (Bit Manipulation - XOR):**
XORing a number with itself results in 0. By XORing all indices from `0` to `n` and all numbers in `nums`, the pairs will cancel out, leaving only the missing number.

```python
class Solution:
    def missingNumber(self, nums: List[int]) -> int:
        missing = len(nums)
        for i, num in enumerate(nums):
            missing ^= i ^ num
        return missing
```

**Complexity Comparison:**
* **Your Solution**: O(N) Time, O(N) Space (for the set).
* **Mathematical Solution**: O(N) Time (for `sum()`), O(1) Space.
* **XOR Solution**: O(N) Time, O(1) Space.

```python
class Solution:
    def countPrimes(self, n: int) -> int:
        if n <= 2:
            return 0
        
        # Initialize a boolean array 'is_prime' with True
        is_prime = [True] * n
        is_prime[0] = is_prime[1] = False

        for i in range(2, int(n**0.5) + 1):

        return sum(is_prime)
```

## Notes: 'Largest Perimeter Triangle' Problem

**Problem Statement:** Given an integer array `nums`, return the largest perimeter of a triangle with a non-zero area, formed from three of these lengths. If it is impossible, return 0.

**The Triangle Inequality Theorem:**
For three lengths to form a triangle, the sum of the two smaller sides must be strictly greater than the longest side ($a + b > c$).

**Optimal Solution (Greedy + Sorting):**
To find the *largest* perimeter, we should sort the array in descending order and check the first triplet that satisfies the triangle condition. If the largest side `nums[i]` cannot form a triangle with the two next largest sides `nums[i+1]` and `nums[i+2]`, it certainly cannot form one with any smaller sides.

```python
class Solution:
    def largestPerimeter(self, nums: List[int]) -> int:
        # Step 1: Sort in descending order
        nums.sort(reverse=True)
        
        # Step 2: Check triplets (i, i+1, i+2)
        for i in range(len(nums) - 2):
            a = nums[i+2]
            b = nums[i+1]
            c = nums[i] # Longest side of the triplet
            
            if a + b > c:
                return a + b + c
                
        return 0
```

**Why this is optimal:**
1.  **Sorting**: By sorting, we ensure we check the largest possible values first.
2.  **Greedy Choice**: If $nums[i+1] + nums[i+2] \le nums[i]$, then any other combination involving $nums[i]$ and even smaller numbers will also fail ($nums[i+1] + nums[i+3]$ will be even smaller).

**Complexity:**
*   **Time Complexity**: O(N log N) due to the sorting step.
*   **Space Complexity**: O(1) or O(N) depending on the sorting implementation's space requirements.

## Notes: 'Sort an Array' Problem

**Problem Statement:** Given an array of integers `nums`, sort the array in ascending order in O(n log n) time and with the smallest space complexity possible, without using built-in functions.

**Your Solution (Merge Sort):**
Your solution uses the divide-and-conquer Merge Sort algorithm. This is a very reliable O(N log N) sorting method.

```python
class Solution:
    def sortArray(self, nums: List[int]) -> List[int]:
        
        def merge(left, right):
            i, j = 0, 0
            result = []
            
            while i < len(left) and j < len(right):
                if left[i] <= right[j]:
                    result.append(left[i])
                    i += 1
                else:
                    result.append(right[j])
                    j += 1
            result.extend(left[i:])
            result.extend(right[j:])
            
            return result
        
        def merge_sort(nums):
            if len(nums) <= 1:
                return nums
            middle = len(nums) // 2
            left = merge_sort(nums[:middle])
            right = merge_sort(nums[middle:])

            return merge(left, right)
        
        return merge_sort(nums)
```

**Suggested Optimal Solution (Heap Sort):**
While Merge Sort is O(N log N) time, it requires **O(N)** extra space for the temporary arrays. To achieve the **'smallest space complexity possible'** (O(1)), Heap Sort is the ideal choice as it sorts the array entirely in-place.

```python
class Solution:
    def sortArray(self, nums: List[int]) -> List[int]:
        def heapify(n, i):
            largest = i
            left = 2 * i + 1
            right = 2 * i + 2
            
            if left < n and nums[left] > nums[largest]:
                largest = left
            if right < n and nums[right] > nums[largest]:
                largest = right
            
            if largest != i:
                nums[i], nums[largest] = nums[largest], nums[i]
                heapify(n, largest)

        n = len(nums)
        # Build max heap
        for i in range(n // 2 - 1, -1, -1):
            heapify(n, i)
            
        # Extract elements from heap
        for i in range(n - 1, 0, -1):
            nums[i], nums[0] = nums[0], nums[i]
            heapify(i, 0)
            
        return nums
```

**Complexity Comparison:**
* **Merge Sort**: O(N log N) Time, **O(N)** Space.
* **Heap Sort**: O(N log N) Time, **O(1)** Space.
* **Quick Sort**: O(N log N) Average Time, O(log N) Space, but O(N²) worst-case time.

```python
        counter = 0
        
        # Fill arr1 with elements from arr2 in order
        for j in arr2:
            for k in range(mapping[j]):
                arr1[counter] = j
                counter += 1
        
        # Append the sorted missing elements
        b = 0
        for a in range(counter, len(arr1)):
            arr1[a] = missing[b]
            b += 1
        return arr1
```

**Your Solution 2 (Dictionary + List Comprehension):**
This version uses a set for fast lookups, a dictionary for frequency counting, and list concatenation with comprehensions to assemble the result.

```python
class Solution:
    def relativeSortArray(self, arr1: List[int], arr2: List[int]) -> List[int]:
        # 1. Create a set for O(1) lookups and a dictionary for counts
        arr2_set = set(arr2)
        extra = []
        arr1_dict = dict()

        # 2. Map frequencies and identify 'extra' elements not in arr2
        for i in arr1:
            arr1_dict[i] = arr1_dict.get(i, 0) + 1
            if i not in arr2_set:
                extra.append(i)
        
        # 3. Build the relative part of the array based on arr2 order
        arr = []
        for j in arr2:
            # Use list comprehension to repeat element 'j' by its frequency
            arr += [j for _ in range(arr1_dict[j])]
        
        # 4. Append the sorted 'extra' elements to the end
        arr += sorted(extra)
        
        return arr
```

**Optimized / Cleaner Solution (Counting Sort Approach):**

Since the range of numbers is small (0-1000), we can use the `Counter` class's properties to build the result array directly.

```python
from collections import Counter

class Solution:
    def relativeSortArray(self, arr1: List[int], arr2: List[int]) -> List[int]:
        counts = Counter(arr1)
        res = []
        
        # 1. Add elements that are in arr2 in the correct order
        for num in arr2:
            res.extend([num] * counts.pop(num))
            
        # 2. The remaining elements in 'counts' are those not in arr2
        # Sort the remaining keys and add them
        for num in sorted(counts.keys()):
            res.extend([num] * counts[num])
            
        return res
```

**Key Improvements in the Optimized Version:**
*   **`counts.pop(num)`**: Retrieves the count and removes the key in one step, leaving only the "extra" elements.
*   **`res.extend([num] * count)`**: A more Pythonic way to add multiple copies of an element than nested loops.
*   **Space/Time Efficiency**: All versions run in O(N log N) due to the sorting of extra elements, with O(N) space complexity.

## Notes: 'Group Anagrams' Problem

**Problem Statement:** Given an array of strings `strs`, group the anagrams together. You can return the answer in any order.

**Your Solution 1 (Nested Loops + Sorted List):**
This approach sorts every string first, then uses nested loops and a `visited` set to group matching sorted versions together.

```python
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        result = []
        strs_2 = []

        # Create a list of sorted characters for each string
        for i in range(len(strs)):
            strs_2.append(sorted(strs[i]))

        visited = set()

        for j in range(len(strs_2)):
            if j in visited:
                continue

            temp = []
            # Compare sorted version of j with all others
            for k in range(len(strs_2)):
                if strs_2[j] == strs_2[k]:
                    temp.append(strs[k])
                    visited.add(k)

            result.append(temp)

        return result
```

**Your Solution 2 (Optimized with `defaultdict`):**
This is a significantly more efficient approach. By using the sorted string as a key in a dictionary, you can group anagrams in a single pass.

```python
from collections import defaultdict

class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        # Use a list as the default value factory
        mapping = defaultdict(list)

        for s in strs:
            # Sort string to use as a unique key for all its anagrams
            key = ''.join(sorted(s))
            mapping[key].append(s)
        
        return list(mapping.values())
```

**Complexity Comparison:**
*   **Solution 1**: O(N^2 * K log K) where N is number of strings and K is max length of a string (due to nested loops and sorting).
*   **Solution 2**: O(N * K log K). This is the standard optimal approach because we only iterate through the list once.

## Notes: 'Top K Frequent Elements' Problem

**Problem Statement:** Given an integer array `nums` and an integer `k`, return the `k` most frequent elements.

**Your Solution (using `Counter.most_common`):**
This approach uses the built-in `most_common()` method of the `Counter` class, which is highly readable and specifically designed for this purpose.

```python
from collections import Counter
from typing import List

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        freq = Counter(nums)
        result = []

        # most_common(k) returns a list of the n most common elements
        # and their counts from the most common to the least.
        for num, count in freq.most_common(k):
            result.append(num)

        return result
```

**Optimal Solution (Counter + Heap):**
To find the top `k` items efficiently, we first count frequencies and then use a Min-Heap of size `k` or Python's `heapq.nlargest` function.

```python
from collections import Counter
import heapq

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        # 1. Build frequency map - O(N)
        count = Counter(nums)
        
        # 2. Use heapq.nlargest to get k elements - O(N log k)
        # It uses a heap under the hood to maintain the top k elements efficiently
        return heapq.nlargest(k, count.keys(), key=count.get)
```

**Alternative Solution (Bucket Sort):**
If you want a strictly O(N) time solution, you can use bucket sort where the index of the bucket represents the frequency.

```python
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        count = Counter(nums)
        # Frequency can be at most len(nums)
        buckets = [[] for _ in range(len(nums) + 1)]
        
        for num, freq in count.items():
            buckets[freq].append(num)
            
        res = []
        # Iterate backwards from highest frequency bucket
        for i in range(len(buckets) - 1, 0, -1):
            for num in buckets[i]:
                res.append(num)
                if len(res) == k:
                    return res
```

**Key Takeaways:**
*   **`Counter.most_common(k)`**: Internally uses `heapq.nlargest` if `k` is small, making it very efficient.
*   **`heapq.nlargest`**: Great for general 'Top K' problems where you have a custom sorting key.
*   **Bucket Sort**: Achieves linear time O(N) by using the frequency as an index.

## Notes: 'Sort Characters By Frequency' Problem

**Problem Statement:** Given a string `s`, sort it in decreasing order based on the frequency of the characters.

**Your Solution (using `Counter.most_common`):**
This approach leverages the `most_common()` method to get characters in descending order of frequency and repeats them using the `*` operator.

```python
from collections import Counter

class Solution:
    def frequencySort(self, s: str) -> str:
        freq = Counter(s)
        result = ""
        # most_common() returns (char, count) tuples sorted by count
        for key, values in freq.most_common(len(s)):
                result += key * values
        return result
```

**Optimized / Pythonic Solution (using `.join()`):**
In Python, strings are immutable, so `result += ...` creates a new string in every iteration. Using a list comprehension with `''.join()` is generally faster (O(N)) because it allocates memory for the final string only once.

```python
from collections import Counter

class Solution:
    def frequencySort(self, s: str) -> str:
        counts = Counter(s)
        
        # 1. Get characters sorted by frequency
        # 2. Multiply character by its count
        # 3. Join the list into a single string
        return "".join([char * count for char, count in counts.most_common()])
```

**Key Takeaways:**
*   **String Concatenation**: Prefer `''.join(list_of_strings)` over `+=` inside loops for better performance in large inputs.
*   **`most_common()`**: Without an argument, it returns all elements sorted by frequency, which is perfect for this problem.

**Complexity:**
*   **Time Complexity**: O(N log K) or O(N + K log K) where N is the length of the string and K is the number of unique characters. Sorting K unique characters takes O(K log K).
*   **Space Complexity**: O(N) to store the result string and the frequency map.
